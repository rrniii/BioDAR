```python
import matplotlib.pyplot as plt
import xarray as xr
import metpy
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from netCDF4 import Dataset
```

```python
nc_fid ='/Volumes/Neely/GREENLAND/ICECAPSarchive/ERA_5/73N_39_5W_72N_35_5W/raw_data/surface/era5_surface_Summit_19790126.nc'
nc4ds=Dataset(nc_fid)
ds = xr.open_dataset(nc_fid,)
data_var = ds.metpy.parse_cf('t2m')
```

```python
mean_t2m=data_var.mean(dim="latitude").mean(dim="longitude")
```

```python
fig = plt.figure(figsize=[20,15], facecolor='white')

plt.plot(data_var.time,mean_t2m-273.15)
plt.show()
```
