```python
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

### Chen et al 2020 Pure Chitin NanoSheets ###


```python
d=10
data = pd.read_csv('/Users/rrniii/University of Leeds/Christopher Hassall - BioDAR Project Folder/WP1 - Electromagnetic (EM) modelling/WP1b - Electromagnetic simulations/DielectricConstants/Chen_et_al_2020_DielectricConstant.csv')  # load data set
X = np.log10(data.iloc[:d, 0].values.reshape(-1, 1)) # values converts it into a numpy array
Y = data.iloc[:d, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
X_pred=np.arange(1,12).reshape(-1, 1)
Y_pred = linear_regressor.predict(X_pred)
X_pred_Xband=np.log10(9.4*10**9).reshape(-1, 1)
Y_pred_Xband = linear_regressor.predict(X_pred_Xband)
```

```python
X_plot = np.log10(data.iloc[:, 0].values.reshape(-1, 1)) # values converts it into a numpy array
Y_plot = (data.iloc[:, 1].values).reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column

plt.plot(X_plot, Y_plot, 'og')
plt.plot(X, Y,'om')
plt.plot(X_pred_Xband, Y_pred_Xband,'xk')
plt.plot(X_pred, Y_pred, color='red')
plt.title('Chen et al. Dielectric Constant Linear Fit')
plt.xlabel('log10(Frequency)')
plt.ylabel('Dielectric Constant')
plt.show()
print('\u03B5\' is ', Y_pred_Xband.item())
```

```python
linear_regressor.score(X, Y) 
```

```python
from sklearn.preprocessing import PolynomialFeatures
X = np.log10(data.iloc[:, 0].values.reshape(-1, 1)) # values converts it into a numpy array
Y = (data.iloc[:, 1].values).reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column


poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, Y)

# Visualizing the Polymonial Regression results
def viz_polymonial():
    plt.scatter(X, Y, color='m')
    plt.plot(X_pred, pol_reg.predict(poly_reg.fit_transform(X_pred)), color='r')
    plt.title('Chen et al. Dielectric Constant Polynomial Fit')
    plt.xlabel('log10(Frequency)')
    plt.ylabel('Dielectric Constant')
    plt.plot(X_pred_Xband,pol_reg.predict(poly_reg.fit_transform(X_pred_Xband)),'xk')
    plt.show()
    return
viz_polymonial()
print('\u03B5\' is ', pol_reg.predict(poly_reg.fit_transform(X_pred_Xband)).item())
```

```python
pol_reg.score(X_poly, Y) 
```
